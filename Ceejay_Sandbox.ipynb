```python
import matplotlib.pyplot as plt
import numpy as np
import random
#import seaborn as sns
```

```python
length_gene=50
n_amino_acids=20
```

```python
#code to get a random array of amino acids that occur at each locus (aas = amino acids)
aas=np.zeros((n_amino_acids,length_gene),dtype=int) 
likely_hood_array=np.random.rand(n_amino_acids)/2  
for i in range(length_gene):
    for j in range(n_amino_acids):
        if random.random() < likely_hood_array[j]:
            aas[j,i]=1
```

```python
#how many different amino acids present at each locus)
n_variants=aas.sum(axis=0) 
```

```python
#plotting
h_ratios=np.ones((n_amino_acids+1),dtype=int) #controlling the relative size of the first subplot
h_ratios[0]=10 # first chart (bar chart) is 10 times the height of the others

#create figures and axes
fig, ax = plt.subplots(n_amino_acids+1,1, gridspec_kw = {'height_ratios':h_ratios}) #creating n_aminoacids number of figures 
fig.set_size_inches(20,10)

#bar chart is showing number of variants at each locus
ax[0].bar(range(len(n_variants)),n_variants, width=1, align='edge')
ax[0].margins(tight=True, x=0)
ax[0].set_xticks(())

#"tick chart" showing which amino acids are present (gray) in each locus
for i in range (n_amino_acids):
    ax[i+1].set_xticks(())
    ax[i+1].set_yticks(())
    ax[i+1].imshow(np.tile(aas[i,:],(1,1)),aspect='auto', cmap='Accent')
plt.show()
```
