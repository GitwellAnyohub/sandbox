```python
import matplotlib.pyplot as plt
```

```python
image_1=([8,2],[4,3])

image_2=([4,3],[7,6])

image_3=([1,1],[13,4])

image_4=([2,1],[4,5])

images=[image_1,image_2,image_3,image_4]
```

```python
w=10
h=10
fig=plt.figure(figsize=(8, 8))
columns = 2
rows = 2
for i in range(1, columns*rows +1):
    img = images[i-1]
    fig.add_subplot(rows, columns, i)
    plt.imshow(img)
plt.show()
```
